```python
def missing_check(df):
    missing = df.isnull().sum()
    missing_per = round(missing/len(df),4)*100
    unique_val = df.nunique()
    type_data = df.dtypes
    df = pd.DataFrame({'Missing_values':missing,
                      'Percent of Missing (%)':missing_per,
                      'Numbers of Unique':unique_val,
                      'Data Type':type_data}).sort_values('Percent of Missing (%)', ascending=False)
    return df
```

```python
missing_check(raw_data)
```

# Encoding

```python
df = raw_data.copy()
```

```python
onehot_enc = ['enginelocation', 'drivewheel', 'doornumber', 'aspiration', 'fueltype']
label_enc = ['fuelsystem', 'cylindernumber', 'enginetype', 'carbody']
```

```python
# Label Encoding
#df[label_enc] = df[label_enc].apply(preprocessing.LabelEncoder().fit_transform)
# onehot
df = pd.get_dummies(df, columns=onehot_enc+label_enc)
```

```python
df.corr()['price'].sort_values(ascending=False)
```

```python
df
```

# Split Data

```python
x = df.drop('price', axis=1)
y = df['price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
y_test.shape
```

# Model Training 1

use model
- Gradient boosting 
- random forest 
- adaboost
- extra trees

### Gradient Boosting

```python
gbr = GradientBoostingRegressor(random_state=0)
gbr.fit(x_train, y_train)
```

```python
test_data_prediction_gbr = gbr.predict(x_test)

r2_test = r2_score(y_test, test_data_prediction_gbr)
mae_test = mean_absolute_error(y_test, test_data_prediction_gbr)
mse_test = mean_squared_error(y_test, test_data_prediction_gbr)
print('R Squared value :' , r2_test)
print('MAE :' , mae_test)
print('MSE :' , mse_test)
```

### Random Forest

```python
rfr = RandomForestRegressor(max_depth=2, random_state=0)
rfr.fit(x_train, y_train)
```

```python
test_data_prediction_rfr = rfr.predict(x_test)

r2_test = r2_score(y_test, test_data_prediction_rfr)
mae_test = mean_absolute_error(y_test, test_data_prediction_rfr)
mse_test = mean_squared_error(y_test, test_data_prediction_rfr)
print('R Squared value :' , r2_test)
print('MAE :' , mae_test)
print('MSE :' , mse_test)
```

### AdaBoost

```python
abr = AdaBoostRegressor(random_state=0)
abr.fit(x_train, y_train)
```

```python
test_data_prediction_abr = abr.predict(x_test)

r2_test = r2_score(y_test, test_data_prediction_abr)
mae_test = mean_absolute_error(y_test, test_data_prediction_abr)
mse_test = mean_squared_error(y_test, test_data_prediction_abr)
print('R Squared value :' , r2_test)
print('MAE :' , mae_test)
print('MSE :' , mse_test)
```

### Extra Trees

```python
etr = ExtraTreesRegressor(random_state=0)
etr.fit(x_train,y_train)
```

```python
test_data_prediction_etr = etr.predict(x_test)

r2_test = r2_score(y_test, test_data_prediction_etr)
mae_test = mean_absolute_error(y_test, test_data_prediction_etr)
mse_test = mean_squared_error(y_test, test_data_prediction_etr)
print('R Squared value :' , r2_test)
print('MAE :' , mae_test)
print('MSE :' , mse_test)
```

```python
estimator = ExtraTreesRegressor(random_state=0)
selector = RFECV(estimator, step=1, cv=5)
selector = selector.fit(x_train, y_train)
```

```python
print('Optimum number features: %d'% selector.n_features_)
```

```python
features = x_train.columns.to_list()
for x,y in (sorted(zip(selector.ranking_, features), key=itemgetter(0))):
    print(x,y)
```

### Feature Selection

```python
feature_selection = ['wheelbase', 'carlength', 'carwidth', 'curbweight', 'enginesize',
                    'boreratio', 'horsepower', 'peakrpm', 'citympg', 'highwaympg', 'drivewheel',
                    'aspiration', 'fuelsystem','cylindernumber', 'price']
```

```python
dataset = raw_data[feature_selection].copy()
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset, columns=['drivewheel', 'aspiration', 'fuelsystem', 'cylindernumber'])
```

```python
dataset
```

```python
x = dataset.drop('price', axis=1)
y = dataset['price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
etr = ExtraTreesRegressor(random_state=0)
etr.fit(x_train,y_train)

test_data_prediction_etr = etr.predict(x_test)

r2_test = r2_score(y_test, test_data_prediction_etr)
mae_test = mean_absolute_error(y_test, test_data_prediction_etr)
mse_test = mean_squared_error(y_test, test_data_prediction_etr)
print('R Squared value :' , r2_test)
print('MAE :' , mae_test)
print('MSE :' , mse_test)
```

# Hyperparameter Tuning

```python
n_estimators = [500,1000,1500,2000,2500,3000]
criterion = ['squared_error', 'absolute_error']
max_depth = [10,15,20,25,30,35,40,45,50]
bootstrap = [True, False]
```

```python
hyperF = dict(bootstrap = bootstrap, n_estimators = n_estimators, criterion = criterion,
              max_depth =max_depth)

gridF = RandomizedSearchCV(ExtraTreesRegressor(random_state=1), hyperF, cv=5, verbose=1,
                           n_jobs=1)
bestF = gridF.fit(x_train, y_train)
```

```python
gridF.best_estimator_
```

```python
etr = ExtraTreesRegressor(bootstrap=True, criterion='absolute_error', max_depth=10,
                    n_estimators=500, random_state=1)
etr.fit(x_train,y_train)
```
